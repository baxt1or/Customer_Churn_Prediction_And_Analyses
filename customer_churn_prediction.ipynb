```python
import pandas as pd
```

```python
customer_df = pd.read_csv("./data/customer_churn.csv")

customer_df.head()
```

```python
customer_df.shape
```

```python
customer_df.columns
```

```python
customer_df["Churn"].value_counts()
```

```python
customer_df.columns = customer_df.columns.str.lower()

customer_df.head()
```

```python
customer_df["gender"] = customer_df["gender"].apply(lambda x: 1 if x == 'Male' else 0)

binary_cols = ["partner", "dependents", "onlinebackup", "churn","phoneservice","onlinesecurity", "deviceprotection", "techsupport", "streamingtv", "streamingmovies", "paperlessbilling"]

for col in binary_cols:
    customer_df[col] = customer_df[col].apply(lambda x: 1 if x == "Yes" else 0)


customer_df.head()
```

```python
customer_df["multiplelines"] = customer_df["monthlycharges"].apply(lambda x: 1 if x == "Yes" else 0)

customer_df
```

```python
customer_df["contract"].unique()

customer_df["contract"] = customer_df["contract"].apply(lambda x: 31 if x == 'Month-to-month' else 365 if x == 'One year' else 365*2 if x == 'Two year' else x)

customer_df.head()
```

```python
string_cols = customer_df.select_dtypes(include="object")

customer_df["internetservice"].unique()
```

```python
customer_df["internetservice"] = customer_df["internetservice"].apply(lambda x: 1 if x == 'DSL' else 2 if x == 'Fiber optic' else 0)

customer_df.head()
```

```python
customer_df["paymentmethod"].unique()
```

```python
customer_df["paymentmethod"] = customer_df["paymentmethod"].apply(lambda x: 1 if x == 'Electronic check' else 2 if x == 'Mailed check' else 3 if x == 'Bank transfer (automatic)' else 4)

customer_df.head()
```

```python
customer_df["totalcharges"] = customer_df["totalcharges"].str.replace(' ', '0').astype(float)

customer_df.head()
```

```python
customer_df.describe().T
```

```python
customer_df.corr(numeric_only=True)["churn"]
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from imblearn.over_sampling import SMOTE

scaler = StandardScaler()

smote = SMOTE(random_state=42)

X_scal = scaler.fit_transform(customer_df.drop(["customerid","churn"], axis=1))
y_scal = customer_df["churn"]

X, y = smote.fit_resample(X_scal,y_scal)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, precision_score, confusion_matrix, recall_score

model = RandomForestClassifier(random_state=42)

model.fit(X_train, y_train)
y_pred = model.predict(X_test)

print("Accuracy: ", accuracy_score(y_test, y_pred))
print("Precision: ", precision_score(y_test, y_pred))
print("Recall: ", recall_score(y_test, y_pred))
print("Confusion matrix:\n", confusion_matrix(y_test, y_pred))
```
